### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
chipo
```

### Step 4. How many products cost more than $10.00?

```python
func = lambda x: float(x[1:-1])
chipo["item_price"] = chipo["item_price"].apply(func)
```

```python
chipo = chipo.drop_duplicates(subset="item_name")
newchipo = chipo[chipo["item_price"]>10.0]
newchipo
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
newchipo = chipo[["item_name","item_price"]]
newchipo = newchipo.sort_values("item_price",ascending=False)
newchipo
```

### Step 6. Sort by the name of the item

```python
newchipo.sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")

#chipo.groupby("item_name").sort_values("item_price",ascending=False)
#chipo[["item_name","item_price"]].sort_values("item_price",ascending=False)
```

```python
most_expensive_item = chipo[["item_name","quantity"]][chipo["item_price"]==chipo["item_price"].max()]
most_expensive_item.groupby("item_name").sum().sort_values("quantity",ascending=False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
len(chipo[chipo["item_name"] == "Veggie Salad Bowl"])
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
len(chipo[(chipo["item_name"] == "Canned Soda") & (chipo["quantity"]>1) ])
```
